```python
#   good, data ( dropped video games dataset )
# use a gradient descent to find mse

def gradient_descent(m_now,b_now,points,Learning_rate):
  # iterate through all pts, get mse then subtract with m_now x L
  m_gradient = 0
  b_gradient = 0
  n = len(points)

  for i in range(n):
    x = points.iloc[i].exp
    y = points.iloc[i].salary
    # use the formula
    y_pred = m_now * x + b_now
    m_gradient  += -(2/n) * x *(y - y_pred)
    b_gradient  += -(2/n) * (y - y_pred)
  m = m_now - (m_gradient * Learning_rate)
  b = b_now - (b_gradient * Learning_rate)
  return m,b
```

```python
# define the variables
m = 0
b = 0
Learning_rate= 0.0001
epochs = 200
```

```python
# train it for the specified epoch
for i in range(epochs):
  m,b = gradient_descent(m,b,df,Learning_rate)

  # print the epoch rate
  if ( i%10 == 0):
    print(f"epoch {i}: b = {b} , m = {m}")
```

```python
# plot the result
plt.scatter(df["exp"],df["salary"])
plt.plot(df["exp"], m * df["exp"] +b, color = "red")
plt.xlabel("Months of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression")
plt.show()
```
